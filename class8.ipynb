```python
# MATPLOTLIB - Data visualisation technique based on Numpy
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,20,0.2)
y = np.sin(x)
plt.plot(x,y)
plt.fill_between(x,y,color='blue', alpha=0.1)
```

```python
x = np.arange(0,20,0.2) #creating a numpy array
y = np.sin(x)           #taking the sine of all the values in the numpy array
plt.plot(x,y)           #plotting the x,y values on a graph 
plt.fill_between(x,y,color='blue', alpha=0.9) #adding a color shade in between the graph fill , alpha is transparency 
```

```python
x = np.linspace(-2,2,1000)
y = np.exp(x)
plt.scatter(x,y)
```

```python
x = np.linspace(-2,2,1000)
y = np.exp(x)
plt.scatter(x,y, c=y)
```

```python
x = np.linspace(-2,2,1000)
y = np.exp(x)
plt.scatter(x,y, c=x)
plt.colorbar()
```

```python
a = [1,3,5,7,9]
b = (4,6,8,10,12)
plt.plot(a,b)
```

```python
a = [1,3,5,7,9]
b = (4,6,8,10,12)
c = [3,6,5,2,8]
plt.plot(a,b)
plt.plot(a,c)
```

```python
a = [1,3,5,7,9]
b = (4,6,8,10,12)
c = [3,6,5,2,8]
plt.plot(a,b, linewidth = 4, marker = '*', markersize=40, markerfacecolor='red', markeredgecolor='yellow', label='blue', linestyle = 'dotted')
plt.plot(a,c, color='red', linewidth=5, marker=',', label = 'red')
# print the title 
plt.title("LIST")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(title = "Blue Line")
# different linestyles - solid, dotted, dashed, dashdot
```

```python
# BAR GRAPH
x = ['aa', 'bb', 'cc', 'dd', 'ee']
y = [50, 15, 37, 25, 64]
plt.bar(x,y, color='yellow')
plt.xticks(rotation = 45) #rotated the text written at the bottom by 45 degrees
plt.grid() #adds gridlines 
plt.savefig("bar_graph") #bar_graph.png gets saved
```

```python
# HORIZONTAL BAR GRAPH
x = ['aa', 'bb', 'cc', 'dd', 'ee']
y = [50, 15, 37, 25, 64]
plt.barh(x,y, color='yellow')
plt.xticks(rotation = 45) #rotated the text written at the bottom by 45 degrees
plt.grid() #adds gridlines 
plt.savefig("bar_graph") #bar_graph.png gets saved
```
